```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42) # to make this code example reproducible 
m = 100 # number of instances
X = 4 * np.random.rand(m, 1) # column vector
y = 4 + 2 * X + np.random.randn(m, 1)  # column vector
```

```python
from regression import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X, y)
print(lr.theta)
X_test = np.array([[0], [4.2]])

y_pred = lr.predict(X_test)
print(y_pred)

#plt.scatter(X.T, y.T, marker="r")
plt.plot(X, y, "g.")
plt.plot(X_test, y_pred, "r-", label="Predictions")
# plt.scatter(X, y)
plt.grid(True)
plt.show()
```

```python
lr2 = LinearRegression()
lr2.fit(X, y)
print(lr2.theta)
X_test = np.array([[0, 1], [4.2, 3]])

y_pred = lr2.predict(X_test)
print(y_pred)
print(lr2.intercept_, lr2.coef_)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(X[:,0],X[:,1], y, c=y, cmap='Reds')
ax.plot(X_test[:,0], X_test[:,1], y_pred.squeeze(), 'green')
```
