```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42) 
m = 100
X = 6 * np.random.rand(m, 1) - 3
y =0.5 * X ** 2 +  0.5 * X ** 2 + X + 2 + np.random.randn(m, 1)
```

```python
from regression import LinearRegression
plt.plot(X, y, "g.", label="Predictыions")

lr = LinearRegression()
X_poly = np.c_[X, np.power(X, 2)]
lr.fit(X_poly, y)
print(lr.coef_, lr.intercept_)

y_pred = lr.predict(X_poly)
print(X_poly.shape, y_pred.shape)
plt.plot(X, y_pred, "r.", label="Predictions")
#plt.scatter(X.T, y.T, marker="r")

# plt.scatter(X, y)
plt.grid(True)
plt.show()
```
