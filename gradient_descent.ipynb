```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

gradient_x_2=lambda v:  v**2

def gradient_descent(
    gradient, start, learn_rate, n_iter=50, tolerance=1e-06
):
    vector = start
    for _ in range(n_iter):
        plt.plot(vector, gradient_x_2(vector), "ro")
        
        diff = -learn_rate * gradient(vector)
        if np.all(np.abs(diff) <= tolerance):
            break
        plt.plot([vector, vector + diff], [gradient_x_2(vector),gradient_x_2(vector + diff)] , "r-", linewidth=0.7)
        
        vector += diff

    return vector



plt.plot(np.linspace(-10, 10, 100), [gradient_x_2(i) for i in np.linspace(-10, 10, 100)])
print([gradient_x_2(i) for i in np.linspace(-10, 10, 100)])
gradient_descent(gradient=lambda v: 2 * v, start=10.0, learn_rate=0.2)
```

```python
plt.plot(np.linspace(-10, 10, 100), [gradient_x_2(i) for i in np.linspace(-10, 10, 100)])
print([gradient_x_2(i) for i in np.linspace(-10, 10, 100)])
gradient_descent(gradient=lambda v: 2 * v, start=10.0, learn_rate=0.8)
```
